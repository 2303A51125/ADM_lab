```python
import pandas as pd
data = pd.read_csv('/content/titanic_clean.csv')
print(data)
```

```python
summary = data.describe()
print(summary)
```

```python
print(data.describe(include=['object']))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(data['age'].dropna(), bins=20, edgecolor='black', color='green')
plt.title('age Distribution')
plt.xlabel('age')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(6, 4))
data['pclass'].value_counts().plot(kind='bar', color='lightcoral')
plt.title('Passenger Class Distribution')
plt.xlabel('Class')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.boxplot(data['age'].dropna(), patch_artist=True, boxprops=dict(facecolor='lightblue'))
plt.title('Box plot of age')
plt.ylabel('age')
plt.show()
```

```python
import seaborn as sns
```

```python
sns.pairplot(data[['age', 'fare', 'sibsp','parch']].dropna())
plt.show()
```

```python
correlation_matrix = data[['age', 'fare', 'sibsp', 'parch']].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.countplot(x='survived', data=data, palette='Set2')
plt.title('survival Count')
plt.xlabel('survived')
plt.ylabel('Count')
plt.show()
```

```python
sns.violinplot(x='pclass', y='age', data=data, palette='muted')
plt.title('Age Distribution by Passenger Class')
plt.show()
```
